```python
import pandas as pd
import numpy as np

import csv

INPUT_CSV = 'data/UN_VOTING_DATA_RAW.csv'
OUTPUT_CSV = 'data/UN_VOTING_DATA_CLEANED.csv'
EXPECTED_COLUMNS = 192

cleaned_rows = []

with open(INPUT_CSV, 'r', encoding='utf-8') as infile:
    reader = csv.reader(infile)

    for row_num, row in enumerate(reader, start=1):
        if len(row) > EXPECTED_COLUMNS:
            print(f"⚠️ Line {row_num} has too many columns ({len(row)}). Merging extras.")
            # Merge extra columns into the last column
            row = row[:EXPECTED_COLUMNS-1] + [','.join(row[EXPECTED_COLUMNS-1:])]

        elif len(row) < EXPECTED_COLUMNS:
            print(f"⚠️ Line {row_num} has too few columns ({len(row)}). Padding.")
            row += [''] * (EXPECTED_COLUMNS - len(row))

        cleaned_rows.append(row)

# Write cleaned rows to new file
with open(OUTPUT_CSV, 'w', newline='', encoding='utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(cleaned_rows)

print(f"✅ Cleaned file saved as {OUTPUT_CSV}")
```
